```python
def plot_offset(start_angle = -30,idx = 1000):
    #idx = 15292
    angles = np.linspace(0,360,360*2)[:-1]
    #positions = angles[np.argmax(fly.state_estimates,axis=1)]
    positions = angles[np.argmax(fly.correlations,axis=1)]
    #positions = angles[np.argmax(filt_cor,axis=1)]
    first_frame = np.squeeze(np.argwhere(np.array(fly.avepxl)>10)[0]) + 1
    fig = plt.figure()
    ax1 = fig.add_subplot(1,1,1)
    img = ax1.imshow(fly.images[idx+first_frame],cmap = plt.cm.gray)
    ax1.axis('off')
    ax2 = fig.add_subplot(1,1,1,projection = 'polar',frameon = False)
    ax2.patch.set_alpha(0.0)
    ax2.plot([np.deg2rad(positions[idx]+start_angle),np.deg2rad(positions[idx]+start_angle)],[0,0.5],'-o')
    ax2.set_ybound(0,1)

def rewrap(trace,offset = np.pi/2.):
    unwrapped = np.unwrap(trace,np.pi*1.8)
    vel = np.diff(unwrapped)
    return np.mod(unwrapped+np.deg2rad(offset),2*np.pi),vel
angles = np.linspace(0,360,360*2)[:-1]
```

```python
flynum = 1328
fly = flb.NetFly(flynum)
fly.open_signals(['hdf5','txt'])
start_angle = 168
plot_offset(start_angle = start_angle,idx = 5000)
fly.save_txt(str(start_angle),'start_angle')
position_ = np.deg2rad(angles[np.argmax(fly.state_estimates,axis = 1)])
position,velocity = rewrap(position_,offset = start_angle)
fly.save_hdf5(position,'position',overwrite = True)
fly.save_hdf5(velocity,'velocity',overwrite = True)
```
